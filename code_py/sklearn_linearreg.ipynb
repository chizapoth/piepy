# Linear regression

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.RandomState(42)
x = 10 * rng.rand(50) # generate 50 random numbers
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y);
```

```python
# choose a class of model
from sklearn.linear_model import LinearRegression
```

```python
# instantiate LinearRegression class, specify we fit the intercept
model = LinearRegression(fit_intercept=True)
model
```

```python
x # it is a np array
```

```python
# put it as a N by p matrix (even if it's just one feature)
X = x[:, np.newaxis] # this puts it as a matrix
```

```python
X.shape
```

```python
X[0:5]
```

```python
# fit model to data
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# prediction
xfit = np.linspace(-1, 11)
```

```python
xfit
```

```python
# also put in the right format
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
```

```python
plt.scatter(x, y)
plt.plot(xfit, yfit) # add the straight line to the plot
```
